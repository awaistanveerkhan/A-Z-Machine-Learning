```python
import pandas as pd
import matplotlib.pyplot as plt
```

Just for plotting purpose, we're limiting our dataset to just two features so we can easily plot them in 2D plan.

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km = KMeans(n_clusters=5, init='k-means++',random_state=0)
y_kmeans = km.fit_predict(X)
```

```python
print(y_kmeans)
```

```python
plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1],s=50,color='red',label='Cluster-01')
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1],s=50,color='blue',label='Cluster-02')
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1],s=50,color='green',label='Cluster-03')
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1],s=50,color='cyan',label='Cluster-04')
plt.scatter(X[y_kmeans == 4,0],X[y_kmeans == 4,1],s=50,color='magenta',label='Cluster-05')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=300,color='yellow',label='Centroids',marker='*')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend()
plt.show()
```
